```python
from zipfile import ZipFile
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
import re

file_name = "xml_containing_html.xml.zip"

# opening the zip file in READ mode
with ZipFile(file_name, 'r') as zip:
     zip.extractall()
#parsing xml 
mytree= ET.parse('xml_containing_html.xml')
myroot=mytree.getroot()

for Each_child in myroot:

    HTML_Data = Each_child.find('content')#extractig html from xml
    soup = BeautifulSoup(HTML_Data.text, 'html.parser')#parsing html
    for Each_link in soup.find_all('a'):
         print(Each_link.get('href')) #printing each link in a html page

    No_of_links = re.findall(r'href', HTML_Data.text)
    print('Number of links in HTML page is {} '.format(len(No_of_links)))
```
